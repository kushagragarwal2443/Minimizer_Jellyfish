```python
from Bio import SeqIO
import matplotlib.pyplot as plt
```

```python
def plotting(input_file):
    
    fasta_sequences = SeqIO.parse(open(input_file),'fasta')    
    
    distri_dict = dict()
    limit = 200

    for i in range(limit+1):
        distri_dict[i] = 0

    for fasta in fasta_sequences:
        count = int(fasta.id)
        if(count < limit):
            distri_dict[count] +=1
        else:
            distri_dict[limit] +=1

    names = list(distri_dict.keys())
    values = list(distri_dict.values())

    sum = 0
    unique = 0
    for i in range(limit+1):
        sum += values[i]*i
        unique += values[i]

    print("TOTAL NUM OF UNIQUE KMERS:", unique)
    print("TOTAL COUNT OF KMERS:", sum)

    plt.figure(figsize=(20,7))
    plt.bar(range(len(distri_dict)), values)
    plt.show()
```

```python
plotting("50seq/original_dumps.fa")
```
